```python
igralci["leta"] = (igralci[["starost"]] // 10) * 10 
```

```python
igralci["leta"] == 20
```

```python
def razvrsti_po_desetletjih(igralci):
    deset = 0
    dvajset = 0
    trideset = 0 
    stirideset = 0
    for i in range(len(igralci)):
        if igralci["leta"][i] == 10:
            deset += 1  
        elif igralci["leta"][i] == 20:
            dvajset += 1  
        elif igralci["leta"][i] == 30:
            trideset += 1
        elif igralci["leta"][i] == 40:
            stirideset += 1
    return deset, dvajset, trideset, stirideset
```

```python
igralci_desetletje = razvrsti_po_desetletjih(igralci)
razvrsti_po_desetletjih(igralci)
```
