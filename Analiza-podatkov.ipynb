4.Število igralcev glede na starost in povprečje točk glede na leta

```python
igralci["leta"] = (igralci[["starost"]] // 10) * 10 
```

```python
#prešteje igralce glede na desetletja
def razvrsti_po_desetletjih(igralci):
    deset = 0
    dvajset = 0
    trideset = 0 
    stirideset = 0
    for i in range(len(igralci)):
        if igralci["leta"][i] == 10:
            deset += 1  
        elif igralci["leta"][i] == 20:
            dvajset += 1  
        elif igralci["leta"][i] == 30:
            trideset += 1
        elif igralci["leta"][i] == 40:
            stirideset += 1
    return deset, dvajset, trideset, stirideset


razvrsti_po_desetletjih(igralci)
```

```python
igralci.groupby(["leta"])["ime"].count().plot.bar()
```

```python
#povprečje točk glede na desetletja
igralci.groupby(["leta"])["povprecje_tocke"].mean().plot.bar()
```

```python
#število igralcev glede na posamezno leto
podtabela3 = igralci[["ime", "priimek", "starost"]].copy()
podtabela3.sort_values(by="starost", ascending = False, inplace = True)
igralci_starost = podtabela2.groupby(by = "starost")[["ime"]].count()
igralci_starost.plot.bar()
```
